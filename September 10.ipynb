```python
import numpy as np

N = int(input("Input how many nodes you want: "))
r1, r2, r3, r4 = 1, 1, 1, 1 #ohm 
V_plus = 10 #volts; sample value 

def voltage_finder(N, r1, r2, r3, r4):
    #RHS or basically yung constants
    b = np.zeros(N)
    b[0] = V_plus/r4
    b[1] = V_plus/r1

    #N by N matrix
    A = np.zeros((N,N)) #(since halos zero naman talaga lahat nung value,  it's easier to just append he non-zero values)

    node_1 = np.array([1/r2 + 1/r3 + 1/r4, -1/r3, -1/r2]) #node 1
    for i in range(N-3):
        node_1 = np.append(node_1, 0)
   
    A[0] = node_1

    for j in range(1, N):   # nodes 1 to N-1
        if j % 2 == 0:  # even (top rail)
            if j-2 >= 0:
                A[j][j-2] = -1/r1
            A[j][j-1] = -1/r3
            A[j][j] = 2/r1 + 1/r3 + 1/r4
            if j+1 < N:
                A[j][j+1] = -1/r4
            if j+2 < N:
                A[j][j+2] = -1/r1
        else:  # odd (bottom rail)
            if j-2 >= 0:
                A[j][j-2] = -1/r2
            A[j][j-1] = -1/r4
            A[j][j] = 2/r2 + 1/r3 + 1/r4
            if j+1 < N:
                A[j][j+1] = -1/r3
            if j+2 < N:
                A[j][j+2] = -1/r2
    return np.linalg.solve(A,b)

volt_vec = voltage_finder(N,r1,r2,r3,r4)
I_tot = (V_plus - volt_vec[1])/r1
R_eff = V_plus/I_tot

print(R_eff)
```
